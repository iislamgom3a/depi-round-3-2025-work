```python
# Labels & Titles


x = np.arange(50,120)
y = np.arange(1,10)


font1 = {"family" :"serif" , "color": "blue" , "size" : 20}
font2 = {"family" :"serif" , "color": "red" , "size" : 15}

## Title

plt.title("Sales from we start" , fontdict = font1)

## labels
plt.xlabel("Years",fontdict = font2)
plt.ylabel("Sales",fontdict = font2)
plt.plot(y , marker = "o")

## Grid
plt.grid()

plt.show()
```

```python
## Subplots (rows , columns , location)

# plot1 sin :
x = np.arange(0,5,0.1)
y = np.sin(x)

plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel("Sin")
plt.title("plot1")
# plot2 cos :
y = np.cos(x)

plt.subplot(1,2,2)
plt.plot(x,y)

plt.xlabel("Cos")
plt.title("plot2")

# Super Title
plt.suptitle("Sin & Cos")
plt.savefig("sincos.png")
plt.show()
```

```python
import pandas as pd

data = pd.read_csv("/content/meets.csv")

data.head()
```

```python
data.info()
```

```python
data.dropna(inplace =True)
```

```python
# Format Date

data['Date'] = pd.to_datetime(data['Date'])

data['month'] = data['Date'].dt.month
data['year'] = data['Date'].dt.year
data['day'] = data['Date'].dt.day

data.drop('Date' , axis = 1 , inplace = True )
```

```python
data.info()
```
