## Data Cleaning 
**What to clean"**: Null values, Duplicates, Outliers, Wrong format, Redundent columns, 

### null vlaues 

```python
# axis : 1-> column, 0-> row
data.drop("region_2", axis = 1, inplace= True)
```

```python
round(data.isna().sum() / len(data), 3) * 100
```

### Fill null values(Impute)
**HOW**: Using mean, median, forward-fill, backward-fill

```python
# finding groups to Impute the null values with varying values, not a fixed one 
data.groupby(["province", "variety"])['price'].mean()
```

```python
data['province'].nunique()
```

```python
group = data.groupby(["province", "variety"])['price']
data['price'].fillna(group.transform('median'), inplace = True)
```

```python
round(data.isna().sum() / len(data), 3) * 100
```

```python
# redundancy 
data.head()
```

```python
data.drop(["description", "taster_name", "taster_twitter_handle", "title"], axis=1, inplace=True)
```

```python
round(data.isna().sum() / len(data), 3) * 100
```

```python
data.drop("designation", axis=1, inplace=True)
```

```python
data.head()
```

```python
# fill region_1 nulls
group = data.groupby(["points", "price"])['region_1']
data['region_1'].fillna(group.transform(
    lambda x:x.mode().iloc[0] if not pd.isnull(x).all() else x   
), inplace = True)
```

```python
round(data.isna().sum() / len(data), 3) * 100
```
