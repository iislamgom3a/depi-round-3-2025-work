## Mutability vs Immutability

- Immutable objects can’t be changed after it is created as int, float, bool, string, Unicode, tuple.
- Mutable objects can be changed after it is created as list, dict, set and user-defined classes.

```python
value = 1  # integer : immutable but could be updated
print(id(value))
value = 2  # this creates a new object with the same name
print(id(value))  # python reused the first object here
```

```python
li = [1, 2, 3, 4]  # list : mutable object
print(id(li))
li.append("Islam ")  # the same object didn't change
print(li)
print(id(li))
```

- String indexes starting at 0 in the beginning of the string and working their way from -1 at the end

### String Slicing

```python
str = "Islam Gomaa"
print(str[0:4])
print(str[3 : len(str)])  # slicing to the end
print(str[3:])  # another way to slice to the end of the string
# --------- slicing with a step ----------#
print(str[2::2])  # step = 2
print(str[5:2:-1])  # reverse steps
```

```python
### Global and Local Scope
x = "global"  # global


def foo():
    print("x inside: ", x)


foo()
print("x outside", x)
```

```python
### Global and Local Scope
x = "global"  # global


def foo():
    x = "local"
    print("x inside: ", x)


foo()
print("x outside", x)
```

```python
### Global and Local Scope
x = "global"  # global


def foo():
    global x
    x = "changed from the function"
    print("x inside: ", x)


foo()
print("x outside", x)
```

## APIs 

### API Status Code
- 200: Everything went okay, and the result has been returned (if any).
- 301: The server is redirecting you to a different endpoint. This can happen when a company switches domain names, or an endpoint name is - changed.
- 400: The server thinks you made a bad request. This happens when you send incorrect data or make other client-side errors.
- 401: The server thinks you're not authenticated. Many APIs require login credentials, so this happens when you don't send the right - credentials to access an API.
- 403: The resource you're trying to access is forbidden: you don't have the right permissions to see it.
- 404: The resource you tried to access wasn't found on the server.
- 503: The server is not ready to handle the request.

### REST APIs
- REST (Representational State Transfer):
- Architectural style for designing networked applications
- Uses a stateless communication protocol, typically HTTP

#### Key Principles of REST
| Principle            | Explanation                                                                 | Example                                                                 |
|----------------------|-----------------------------------------------------------------------------|-------------------------------------------------------------------------|
| Statelessness         | Server stores no client session; each request is independent                | Every API request includes auth token or user info                      |
| Client-Server         | UI and logic are separated between client and server                        | Mobile app (UI) calls API (logic) to fetch data                         |
| Cacheability          | Responses must indicate if they're cacheable                                | Response header: `Cache-Control: max-age=600`                           |
| Uniform Interface     | Consistent method to access resources (standard verbs and URIs)             | Use of `GET`, `POST`, `PUT`, `DELETE` across all resources              |
| Layered System        | Client doesn’t know if it's connected to the actual server or intermediaries| Request goes through proxy/cache before hitting the API                 |

## OOP 

```python
# define classes
class Car:
    # constructor
    def __init__(self, make, model):
        self.make = make
        self.model = model

    # define methods
    def accelerate(self):
        print("The car is accelerating.....!")


my_car = Car("Honda", "Civic")
print("My car is: ", my_car.make, my_car.model)
my_car.accelerate()
```

```python
# Inheritance
class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        print("sound is on......")


class Dog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)
        self.breed = breed

    def make_sound(self):
        print("wooooooof....")


my_dog = Dog("budddy", "Labrador")
print(my_dog.name, my_dog.breed)
my_dog.make_sound()
```

```python
# Polymorphism
class Animal:
    def __init__(self, name):
        self.name = name

    def make_sound(self):
        pass


class Dog(Animal):
    def make_sound(self):
        print("Woof!")


class Cat(Animal):
    def make_sound(self):
        print("Meow!")


def make_animal_sound(animal):
    animal.make_sound()


myDog = Dog("Buddy")
myCat = Cat("Whiskers")
make_animal_sound(myDog)
make_animal_sound(myCat)
```
