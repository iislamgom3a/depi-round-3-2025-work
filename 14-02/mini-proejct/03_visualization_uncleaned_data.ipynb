```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("AgeDataset-V1.csv")
```

```python
df.head()
```

```python
sns.histplot(df["Birth year"], bins=30, kde=True)
plt.title("Distribution of Birth Year")
plt.xlabel("Birth Year")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(df["Age of death"], bins=10, kde=True)
plt.title("Distribution of Age of Death")
plt.xlabel("Age of Death")
plt.ylabel("Count")
plt.show()
```

```python
df["Country"].value_counts().head(10).plot(kind="bar")
plt.title("top 10 countres")
plt.xlabel("country")
plt.ylabel("count")
plt.show()
```

```python
df.isnull().sum().plot(kind="bar")
plt.title("missing values")
plt.ylabel("missing count")
plt.show()
```

```python
df["Manner of death"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Distribution of Manner of Death")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Corrlation Heatmap")
plt.show()
```

```python
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.show()
```
